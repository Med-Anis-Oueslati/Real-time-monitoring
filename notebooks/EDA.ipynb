```python
from kafka import KafkaConsumer
import json
import os

# Kafka settings
bootstrap_servers = '192.168.56.1:9093'
topic = 'lubuntu_auth'
output_dir = './lubuntu_auth_data'
output_file = f'{output_dir}/lubuntu_auth_messages.jsonl'
batch_size = 100  # Adjust batch size based on your needs

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Create consumer
consumer = KafkaConsumer(
    topic,
    bootstrap_servers=bootstrap_servers,
    auto_offset_reset='latest',  # Start from the latest message
    enable_auto_commit=True,
    group_id='lubuntu-auth-consumer',
    value_deserializer=lambda x: x.decode('utf-8')  # Assuming UTF-8 messages
)

# Consume and save messages in batches
print(f"Consuming messages from {topic} and saving to {output_file}...")
messages = []
try:
    with open(output_file, 'w') as f:
        for message in consumer:
            print(f"Message: {message.value}")  # Ensure message is being received
            f.write(message.value + '\n')
            print(f"Offset: {message.offset}, Value: {message.value}")  # Debugging offset and message

except Exception as e:
    print(f"Error: {e}")
finally:
    consumer.close()

```
